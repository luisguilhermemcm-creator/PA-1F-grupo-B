```python
# Operedores aritmétricos

# Pedir para o usuário digitar 02 números

#Entrada

a = int(input("Digite um número: "))
b = int(input("Dígite um número: "))

#Processamento

#soma

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b

#Saída

print (f"")
print (f"--------------Operadores arítmértricos----------------")
print (f"")
print (f"------------------------SOMA--------------------------")
print (f"")
print (f"O resultado da soma de {a} mais {b} é igual a : {soma}")
print (f"")
print (f"------------------------------------------------------")
print (f"")
print (f"----------------------SUBTRAÇÃO-----------------------")
print (f"")
print (f"O resultado da subtração de {a} menos {b} é igual a : {subtracao}")
print (f"")
print (f"------------------------------------------------------")
print (f"")
print (f"--------------------MULTIPLICAÇÂO---------------------")
print (f"")
print (f"O resultado da multiplicação de {a} vezes {b} é igual a : {multiplicacao}")
print (f"")
print (f"------------------------------------------------------")
print (f"")
print (f"-----------------------DIVISÃO------------------------")
print (f"")
print (f"O resultado da divisão de {a} dividido por {b} é igual a : {divisao}")
print (f"")
print (f"------------------------------------------------------")
print (f"-----------------------DIVISÃO------------------------")
print (f"")
print (f"O resultado da divisão inteira de {a} dividido por {b} é igual a : {div_int}")
print (f"")
print (f"------------------------------------------------------")
print (f"-----------------------DIVISÃO------------------------")
print (f"")
print (f"O resultado da divisão inteira de {a} dividido por {b} é igual a : {div_int}")
print (f"")
print (f"------------------------------------------------------")
print (f"-----------------------POTÊNCIA------------------------")
print (f"")
print (f"O resultado da potência de {a} elevado a {b} é igual a : {potencia }")
print (f"")
print (f"------------------------------------------------------")

```
