Fazer um programa que o usuario irá entrar com 04 notas , vamos calcular a média e mostrar o usuario a média

```python
#entrada

numero1 = float (input("digite o primeiro numero "))
numero2 = float (input("digite o segundo numero "))
numero3 = float (input("digite o terceiro numero "))
numero4 = float (input("digite o quarto numero "))

#processamento

Media = numero1 + numero2 + numero3 + numero4
resultado = Media/4

#saida

print (f"a media dos 4 numeros é {resultado}")
```
